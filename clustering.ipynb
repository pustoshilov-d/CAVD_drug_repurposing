```python
bioDf = allDF.drop(chemCols, axis=1)
chemDf = allDF.drop(bioCols, axis=1)

summaryDf = initDf[['Source.MineID', 'label', 'Source.DrugBankID']]
summaryDf.columns = summaryDf.columns.str.replace('Source.DrugBankID', 'type')

summaryDf.loc[summaryDf['type'] != 'prime', 'type'] = 'similar' 
summaryDf.columns = summaryDf.columns.str.replace('Source.MineID', 'group')
```

```python
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.decomposition import TruncatedSVD
from sklearn.ensemble import RandomTreesEmbedding, ExtraTreesClassifier

import math


pca = PCA(n_components=1)
pca_bio = pca.fit_transform(bioDf)
pca_bio = pd.DataFrame(pca_bio, columns=['PCA_biological'])
pca_bio.index = pca_bio.index + 1
pca_score = pca.get_covariance
pca_chem = pca.fit_transform(chemDf)
pca_chem = pd.DataFrame(pca_chem, columns=['PCA_chemical'])
pca_chem.index = pca_chem.index + 1

pca_general = pca.fit_transform(allDF)
pca_general = pd.DataFrame(pca_general, columns=['PCA'])
pca_general.index = pca_general.index + 1

pca_average = pd.DataFrame((pca_bio['PCA_biological'] + pca_chem['PCA_chemical']) / 2, columns=['PCA_average'])

summaryDf = summaryDf.join([pca_bio, pca_chem, pca_general, pca_average])

# _________________

lda = LinearDiscriminantAnalysis(n_components=1, solver='svd')

lda_bio = lda.fit_transform(bioDf, summaryDf['group'])
lda_bio = pd.DataFrame(lda_bio, columns=['LDA_svd_biological'])
lda_bio.index = lda_bio.index + 1

lda_chem = lda.fit_transform(chemDf, summaryDf['group'])
lda_chem = pd.DataFrame(lda_chem, columns=['LDA_svd_chemical'])
lda_chem.index = lda_chem.index + 1

lda_general = lda.fit_transform(allDF, summaryDf['group'])
lda_general = pd.DataFrame(lda_general, columns=['LDA_svd'])
lda_general.index = lda_general.index + 1

lda_average = pd.DataFrame((lda_bio['LDA_svd_biological'] + lda_chem['LDA_svd_chemical']) / 2, columns=['LDA_svd_average'])

summaryDf = summaryDf.join([lda_bio, lda_chem, lda_general, lda_average])

# _________________

lda = LinearDiscriminantAnalysis(n_components=1, solver='eigen', shrinkage='auto')

lda_bio = lda.fit_transform(bioDf, summaryDf['group'])
lda_bio = pd.DataFrame(lda_bio, columns=['LDA_eigen_biological'])
lda_bio.index = lda_bio.index + 1

lda_chem = lda.fit_transform(chemDf, summaryDf['group'])
lda_chem = pd.DataFrame(lda_chem, columns=['LDA_eigen_chemical'])
lda_chem.index = lda_chem.index + 1

lda_general = lda.fit_transform(allDF, summaryDf['group'])
lda_general = pd.DataFrame(lda_general, columns=['LDA_eigen'])
lda_general.index = lda_general.index + 1

lda_average = pd.DataFrame((lda_bio['LDA_eigen_biological'] + lda_chem['LDA_eigen_chemical']) / 2, columns=['LDA_eigen_average'])

summaryDf = summaryDf.join([lda_bio, lda_chem, lda_general, lda_average])

# _________________

hasher = RandomTreesEmbedding(n_estimators=10, random_state=0, max_depth=5)
svd = TruncatedSVD(n_components=1, n_iter=20, random_state=42)

svd_bio = hasher.fit_transform(bioDf)
svd_bio = svd.fit_transform(svd_bio)
svd_bio = pd.DataFrame(svd_bio, columns=['SVD_emb_biological'])
svd_bio.index = svd_bio.index + 1

svd_chem = hasher.fit_transform(chemDf)
svd_chem = svd.fit_transform(svd_chem)
svd_chem = pd.DataFrame(svd_chem, columns=['SVD_emb_chemical'])
svd_chem.index = svd_chem.index + 1

svd_general = hasher.fit_transform(allDF)
svd_general = svd.fit_transform(svd_general)
svd_general = pd.DataFrame(svd_general, columns=['SVD_emb'])
svd_general.index = svd_general.index + 1

svd_average = pd.DataFrame((svd_bio['SVD_emb_biological'] + svd_chem['SVD_emb_chemical']) / 2, columns=['SVD_emb_average'])

summaryDf = summaryDf.join([svd_bio, svd_chem, svd_general, svd_average])

# _________________

svd = TruncatedSVD(n_components=1, n_iter=20, random_state=42)

svd_bio = svd.fit_transform(bioDf)
svd_bio = pd.DataFrame(svd_bio, columns=['SVD_biological'])
svd_bio.index = svd_bio.index + 1

svd_chem = svd.fit_transform(chemDf)
svd_chem = pd.DataFrame(svd_chem, columns=['SVD_chemical'])
svd_chem.index = svd_chem.index + 1

svd_general = svd.fit_transform(allDF)
svd_general = pd.DataFrame(svd_general, columns=['SVD'])
svd_general.index = svd_general.index + 1

svd_average = pd.DataFrame((svd_bio['SVD_biological'] + svd_chem['SVD_chemical']) / 2, columns=['SVD_average'])

summaryDf = summaryDf.join([svd_bio, svd_chem, svd_general, svd_average])

# _________________

summaryDf
```

```python
numGroups = len(tempSummaryDf['group'].unique())
meanStd = pd.DataFrame([[]])

for col in tempSummaryDf:
  if not col in set(['label', 'type', 'group']):
    sumStd = 0 
    for group in tempSummaryDf['group'].unique():
      curDf = tempSummaryDf[tempSummaryDf['group'] == group]
      sumStd = sumStd + curDf[col].std()
    meanStd[col] = sumStd / numGroups
meanStd.transpose().sort_values(0)
```

```python
primeDf = summaryDf[summaryDf['type'] =='prime']
primeDf = primeDf.set_index('group')
primeDf
```

```python

nonCols = ['group','label','type']

cols = summaryDf.columns
for columnName in cols:
  if not columnName in set(['group','label','type']):
    new = 'similarityBioChem.'+columnName
    summaryDf[new] = np.nan
    nonCols.append(new)
# cols, nonCols

for rowIndex, row in summaryDf.iterrows():
  print(rowIndex, 'of', summaryDf.shape)
  for columnName, value in row.items():
    group = row['group']
    if not columnName in set(nonCols):
      norm = summaryDf[columnName].max() - summaryDf[columnName].min()
      primeVal = primeDf.loc[group][columnName]
      summaryDf['similarityBioChem.'+columnName][rowIndex] = 1 - (abs(value - primeVal) / norm)
summaryDf
```
