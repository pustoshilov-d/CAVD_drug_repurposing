```python

for i,rowPrime in df.iterrows():
  if rowPrime['type'] == 'prime':
    for j,rowDf in df.iterrows():
      if rowDf['group'] == rowPrime['group']:
        for col in df.columns:
          if 'tar.cur' in col:
            tarName = col.replace('.cur', '')
            if not pd.isna(rowPrime[tarName+'.cur']):
              # print(tarName, rowPrime[tarName+'.cur'])
              df.at[j, tarName+'.ref'] = rowPrime[tarName+'.cur']
              df.at[j, tarName+'.dif'] = df.at[j, tarName+'.ref']  - df.at[j, tarName+'.cur']
df
```

```python
temp = pd.Series()

tarCurCols = [col for col in df if 'tar.cur' in col]
tarDifCols = [col for col in df if 'tar.dif' in col]
tarRefCols = [col for col in df if 'tar.ref' in col]

LIMIT = -9
OFFSET = 1


for i,row in df.iterrows():
  df.at[i, 'hasProccessed'] = 1 if row[tarCurCols].notna().sum() > 0 else 0
  df.at[i, 'numberProccessed'] = row[tarCurCols].notna().sum()
  df.at[i, 'numberRef.limit'] = sum(row[tarRefCols].fillna(1000) < LIMIT)
  df.at[i, 'numberCur.limit'] = sum(row[tarCurCols].fillna(1000) < LIMIT)
  df.at[i, 'numberDif'] = sum(row[tarDifCols].fillna(-1000) > 0 - OFFSET) 

df['meanCur'] = df[tarCurCols].mean(skipna=True, axis=1)
df['meanRef'] = df[tarRefCols].mean(skipna=True, axis=1)
df['meanDif'] = df[tarDifCols].mean(skipna=True, axis=1)
df['procRef'] = df['numberRef.limit'] / df['numberProccessed']
df['procCur'] = df['numberCur.limit'] / df['numberProccessed']

for i,row in df.iterrows():
  res = 0
  if row['numberRef.limit'] != 0:
    if not pd.isna(row['numberRef.limit']):
      # print(row['numberCur.limit'] / row['numberRef.limit'])
      res = row['numberCur.limit'] / row['numberRef.limit']
  df.at[i, 'proc'] = res
  
df

```
